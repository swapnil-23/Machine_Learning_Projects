```python
sns.set(style="darkgrid")
sns.regplot(x = df['year'] ,y= df['mileage'])
```

```python
sns.regplot(x=df['year'] , y=df['mileage'] , marker="*" , fit_reg=False)
```

```python
sns.lmplot(x='year' , y='mileage' ,data = df ,fit_reg=False , hue = 'tax' , legend = True , palette = "Set1")
```

```python
sns.lmplot(x = "year" , y="mileage" , data = df , fit_reg=False , hue = "tax" , legend = True , palette = "Set1")
```

```python
#histogram with fixed number of bins 
sns.distplot(df['mileage'],bins=20)
```

```python
#histogram with fixed number of bins and using kernel density estimate
sns.distplot(df['mileage'] , kde = False , bins = 50)
```

```python
#using barplots
sns.countplot(x = 'fuelType' , data = df , hue = 'transmission')
```

```python
#applying boxplot it helps to interpret the five number summary that includes minimum , maximum  , first, secomd and third quantile
sns.set(style="darkgrid")
sns.boxplot(y=df["mileage"])
```

```python
#numerical vs categorical variable
sns.boxplot(x=df["fuelType"] , y=df['mileage'])
```

```python
#group box and whisker plot
sns.boxplot(x=df["fuelType"] , y=df['mileage'] , hue = "transmission" , data = df)
```

```python
#box-whisker plot and histogram
f,(ax_box , ax_hist) = plt.subplots(2,gridspec_kw={"height_ratios" : (.15 , .85)})

sns.boxplot(df["mileage"] , ax=ax_box)
sns.distplot(df["mileage"] , ax=ax_hist , kde = False)
```

```python
#pairwise plot
sns.pairplot(df , kind = 'scatter' , hue = "fuelType")
plt.show()
```
